# MVP

The goal of this project is to identify the key topics that occur among New York Times' Modern Love column essays and short stories.

After acquiring approximately 1615 Modern Love essays and short stories via the NYT's API and web scraping, I cleaned the dataset and ended up with 1504 essays/short stories (i.e., documents) in the dataset (i.e., corpus). 
Next, I did some initial topic modeling using sklearn's NMF algorithm and the CorEx algorithm.

The below topics and top words associated with them resulted from an NFM model fit on a TF-IDF doc-word matrix with hyperparameters tuned to max_df=.8 and min_df=.01. The corpus was pre-processed to remove punctuation, numbers and some stop words and terms were lemmatized (although, it appears that the lemmatization may not have been successful, since both father, dad and daddy all appear, for example). 

The below topics and top words associated with them resulted from a CorEx model fit on a CountVectorizer doc-word matrix with hyperparameters tuned to max_df=0.8 and min_df=0.003. The corpus was pre-processed to remove punctuation, numbers and some stop words and terms were lemmatized. I also used various anchor words with strength set to 2.

Initially, I thought the topics from the NMF model weren't very distinguishable from one another, which is why I switched over to using the CorEx algorithm. At second glance, though, I think NMF results might be more useful. I see the following topics emerging from them:

- family and family members
- dating and relationships
- marriage and divorce
- family death/loss
- high school

Next, I plan to further pre-process the corpus (i.e., fix lemmatization issues and remove additional stop words) to further improve the NMF model. I also plan to build a K-means clustering model that incorporates word embeddings.
